```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read a preloaded dataset from seaborn
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
```

```python
iris.dtypes
```

```python
#set object type to category
iris['species'] = iris['species'].astype('category')
iris.dtypes
```

# Examples of visualizations

## sns.scatterplot()

```python
plt.figure(figsize=(10, 8))
colors = ['green', 'blue', 'brown']
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species', palette = 'inferno',
    s=100, alpha=0.7)

plt.show()
```

## sns.jointplot()

```python
sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species', palette = 'inferno')
```

```python
sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species', palette = 'inferno', 
    height = 8, ratio = 8, xlim = (0, 9), ylim = (0, 9))
plt.show()
```

```python
sns.jointplot(data=iris[iris['species'] == 'virginica'], x = 'sepal_length', y = 'petal_length', 
    hue = 'species', palette = 'inferno', xlim = (0, 9), ylim = (0, 9))
```

## sns.lmplot()

```python
sns.set_theme()
sns.lmplot(
    data = iris,
    x = 'sepal_width', y = 'petal_width', 
    hue = 'species', palette = 'dark' 
)
```

## sns.pairplot()

https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(data = iris, hue = 'species', palette = 'inferno', 
    diag_kind = 'kde', height = 3, 
    corner = False, kind = 'scatter'
)
```

```python
sns.pairplot(data = iris, hue = 'species', palette = 'inferno', 
    diag_kind = 'kde', height = 2, 
    corner = False, kind = 'kde'
)
```

## sns.boxplot()

```python
# box plot 
sns.boxplot(x = 'species', y = 'sepal_length', data = iris)
```

## sns.barplot()

```python
sns.barplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
iris['sepal_length_means'] = iris.groupby('species')['sepal_length'].transform('mean')
```

```python
sepalLengthMeans = np.array(iris['sepal_length_means'].unique())
sepalLengthMeans
```

Manipulation of seaborn bars at plt level

```python
# box plot 
iris.head()
barPlot = sns.barplot(x = 'species', y = 'sepal_length_means', data = iris, 
    palette = 'inferno', ci = None)

for i in barPlot.containers:
    barPlot.bar_label(i)

plt.show()

### plt.bar()
```

```python
#bar plot of populations of five countries
from tkinter import font


barGraph = plt.bar( iris['species'] , iris['sepal_length_means'], width = 0.5, 
    color = ['red', 'blue', 'green'])
plt.xticks(np.arange(3), iris['species'].unique() )
    #rotation = 45) # x axis config
plt.title('sepal length means by species')
plt.xlabel('Species')
plt.ylabel('sepal length mean')
plt.bar_label(barGraph, fontweight = 'light', font = 'Arial', fontsize = 12)
plt.show()
```

```python
plt.subplot(1, 2, 1)
barplot = plt.bar(iris['species'], iris['sepal_length_means'])
plt.bar_label(barplot, labels = iris['sepal_length_means'])#, padding = 3)
plt.show()
```
