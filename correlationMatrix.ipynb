```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import pandas as pd

iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
sns.pairplot(iris, hue = 'species')
```

# scaling

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled.T
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot( pd.DataFrame( scaled ))
```

```python
scaled.shape
```

```python
sns.displot(scaled, kde = True)
```

```python
scalCopy = scaled.copy()
scalCopy.shape = (150 * 4, 1)
```

```python
sns.displot(scalCopy, kde = True)
```

```python
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
sns.heatmap(covariance_matrix, annot = True)
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.4f',
                 annot_kws={'size': 12},
                 yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
